<h1 style="text-align:center;">:: in R</h1>


<u>**Auteur</u> :** [Younes Dahami](https://www.linkedin.com/in/dahami/)

In R, the **`::`** operator is used to access functions or objects from specific packages without needing to load the entire package namespace. It's particularly useful when we want to use a function from a package without attaching the entire package namespace to your search path.

For example, if we want to use a function called `foo()` from a package named `bar`, we can do so using `bar::foo()`. This ensures that R looks for `foo()` only within the `bar` package, even if there are other functions with the same name in other packages or in your global environment.

Using `::` can help avoid conflicts between function names and reduce clutter in your code, as it makes it clear where each function is coming from.
